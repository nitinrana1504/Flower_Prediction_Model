```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from keras.preprocessing import image
```

```python
model = load_model('my_modelFor128.h5')
```

```python
test_image = image.load_img('FinalTest/calendula.jpeg',target_size=(128,128))
test_image1=image.img_to_array(test_image)
test_image1=np.expand_dims(test_image1,axis=0)
result=model.predict(test_image1)
```

```python
print(result)
```

```python
flower_names = ['Bellflower', 'Calendula', 'Daisy', 'Dandelion', 'Iris', 'Rose', 'Sunflower', 'Tulip']
predicted_flower = flower_names[np.argmax(result)]
print(predicted_flower)
```

```python
plt.imshow(test_image)
plt.title(f'Prediction: {predicted_flower}')
plt.show()
```
